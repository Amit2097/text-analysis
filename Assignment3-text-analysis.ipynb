```python
# Function to preprocess and tokenize text
def preprocess_and_tokenize(text):
    tokenized_story = re.sub(r'[^א-ת\s]', '', text.lower())  # Remove punctuation
    tokens = tokenized_story.split()  # Split text into tokens
    return ' '.join(tokens)
```

```python

# Apply cleaning and tokenization to the 'story' column
df_train['tokenized_story'] = df_train['story'].apply(preprocess_and_tokenize)
df_test['tokenized_story'] = df_test['story'].apply(preprocess_and_tokenize)

# Display the cleaned and tokenized stories
print(df_train[['story', 'tokenized_story']].head())
print(df_test[['story', 'tokenized_story']].head())
```

```python
count_vectorizer = CountVectorizer(max_features=5000)
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Apply vectorization on the tokenized stories
X_train_count = count_vectorizer.fit_transform(df_train['tokenized_story'])
X_test_count = count_vectorizer.transform(df_test['tokenized_story'])

X_train_tfidf = tfidf_vectorizer.fit_transform(df_train['tokenized_story'])
X_test_tfidf = tfidf_vectorizer.transform(df_test['tokenized_story'])

```

```python
from sklearn.model_selection import StratifiedKFold

# Choose the Count Vectorizer
chosen_vectorizer = count_vectorizer
X_train_chosen = chosen_vectorizer.fit_transform(df_train['tokenized_story'])
y_train = df_train['gender']

# Create a list of models you want to try
models = [
    LinearSVC(),
    MLPClassifier(),
    Perceptron(),
    SGDClassifier(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    MultinomialNB(),
    GaussianNB()
]

# Set up cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Create an empty dictionary to store model scores
model_scores = {}

# Iterate through each model
for model in models:
    scores = []  # To store scores for each fold
    for train_idx, val_idx in cv.split(X_train_chosen, y_train):
        X_train_fold, X_val_fold = X_train_chosen[train_idx], X_train_chosen[val_idx]
        y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]
        
        # Create a pipeline with the current model
        if isinstance(model, (MultinomialNB, GaussianNB)):
            # Convert sparse matrix to dense for Naive Bayes models
            X_train_fold = X_train_fold.toarray()
            X_val_fold = X_val_fold.toarray()
        
        pipeline_model = Pipeline([
            ('model', model)
        ])
        
        # Train the model on the current fold
        pipeline_model.fit(X_train_fold, y_train_fold)
        
        # Evaluate the model on the validation fold
        accuracy = pipeline_model.score(X_val_fold, y_val_fold)
        scores.append(accuracy)
    
    # Store the scores for the current model
    model_scores[model.__class__.__name__] = scores

# Print the scores for each model
for model_name, scores in model_scores.items():
    mean_accuracy = sum(scores) / len(scores)
    print(f"{model_name}: Mean Accuracy = {mean_accuracy:.4f}")
```

```python

# Define the hyperparameter grids for each model
param_grids = {
    LinearSVC(): {
        'model__C': [0.1, 1, 10],
        'model__class_weight': ['balanced', None]
    },
    MLPClassifier(): {
        'model__hidden_layer_sizes': [(50,), (100,), (50, 50)],
        'model__activation': ['relu', 'tanh'],
        'model__alpha': [0.0001, 0.001, 0.01]
    },
    Perceptron(): {
        'model__alpha': [0.0001, 0.001, 0.01],
        'model__penalty': ['l1', 'l2', 'elasticnet']
    },
    SGDClassifier(): {
        'model__alpha': [0.0001, 0.001, 0.01],
        'model__penalty': ['l1', 'l2', 'elasticnet'],
        'model__class_weight': ['balanced', None]
    }
}

# Set up cross-validation
cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

# Create a dictionary to store the best models
best_models = {}

# Iterate through each model
for model, param_grid in param_grids.items():
    # Create a pipeline with the current model
    pipeline_model = Pipeline([
        ('model', model)
    ])
    
    # Perform grid search with cross-validation
    grid_search = GridSearchCV(pipeline_model, param_grid, cv=cv, scoring='accuracy')
    grid_search.fit(X_train_chosen, y_train)
    
    # Store the best model and its parameters
    best_models[model.__class__.__name__] = {
        'model': grid_search.best_estimator_,
        'params': grid_search.best_params_,
        'mean_accuracy': grid_search.best_score_
    }

# Print the results for the best models
for model_name, info in best_models.items():
    print(f"{model_name}: Best Parameters = {info['params']}, Mean Accuracy = {info['mean_accuracy']:.4f}")
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import precision_score, recall_score

# Create dictionaries to store precision, recall, and F1 scores for male and female classes
precision_scores = {}
recall_scores = {}
f1_scores = {}

# Set up cross-validation
cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

# Iterate through each best model
for model_name, info in best_models.items():
    model = info['model']
    
    # Initialize lists to store scores for each fold
    male_precision_list = []
    female_precision_list = []
    male_recall_list = []
    female_recall_list = []
    
    # Perform cross-validation predictions and scores
    for train_idx, val_idx in cv.split(X_train_chosen, y_train):
        X_train_fold, X_val_fold = X_train_chosen[train_idx], X_train_chosen[val_idx]
        y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]
        
        model.fit(X_train_fold, y_train_fold)
        y_pred_fold = model.predict(X_val_fold)
        
        # Calculate precision, recall for male and female classes and store in lists
        male_precision_fold = precision_score(y_val_fold, y_pred_fold, pos_label='m')
        female_precision_fold = precision_score(y_val_fold, y_pred_fold, pos_label='f')
        male_recall_fold = recall_score(y_val_fold, y_pred_fold, pos_label='m')
        female_recall_fold = recall_score(y_val_fold, y_pred_fold, pos_label='f')
        
        male_precision_list.append(male_precision_fold)
        female_precision_list.append(female_precision_fold)
        male_recall_list.append(male_recall_fold)
        female_recall_list.append(female_recall_fold)
    
    # Calculate average precision, recall, and F1 scores for male and female classes
    precision_scores[model_name] = {
        'male': sum(male_precision_list) / len(male_precision_list),
        'female': sum(female_precision_list) / len(female_precision_list)
    }
    recall_scores[model_name] = {
        'male': sum(male_recall_list) / len(male_recall_list),
        'female': sum(female_recall_list) / len(female_recall_list)
    }
    f1_scores[model_name] = {
        'male': 2 * (precision_scores[model_name]['male'] * recall_scores[model_name]['male']) /
                (precision_scores[model_name]['male'] + recall_scores[model_name]['male']),
        'female': 2 * (precision_scores[model_name]['female'] * recall_scores[model_name]['female']) /
                  (precision_scores[model_name]['female'] + recall_scores[model_name]['female'])
    }

# Calculate and print Average F1 scores for male and female classes
for model_name in f1_scores.keys():
    avg_f1_male = (f1_scores[model_name]['male'] + f1_scores[model_name]['female']) / 2
    print(f"{model_name}: Average F1 (male) = {f1_scores[model_name]['male']:.4f}, Average F1 (female) = {f1_scores[model_name]['female']:.4f}, Average F1 = {avg_f1_male:.4f}")
```
